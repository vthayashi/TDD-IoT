```python
import pandas as pd

df = pd.read_csv("rawdata-mysql.csv")

df
```

```python
df = df.drop(columns=['ts','topic','chipid','board','firmware'])

df
```

```python
import matplotlib.pyplot as plt

df2 = df[['freeheap', 'wifi_level']].copy()

kmeans = KMeans(n_clusters=3).fit(df2)
centroids = kmeans.cluster_centers_

plt.figure(figsize = (8,8))

plt.scatter(df2['freeheap'], df2['wifi_level'],
            c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)

plt.xlabel("freeheap")
plt.ylabel("wifi_level")

plt.show()
```

```python
df2 = df[['freeheap', 'internal_temperature']].copy()

kmeans = KMeans(n_clusters=3).fit(df2)
centroids = kmeans.cluster_centers_

plt.figure(figsize = (8,8))

plt.scatter(df2['freeheap'], df2['internal_temperature'],
            c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)

plt.xlabel("freeheap")
plt.ylabel("internal_temperature")

plt.show()
```

```python
df2 = df[['wifi_level', 'internal_temperature']].copy()

kmeans = KMeans(n_clusters=2).fit(df2)
centroids = kmeans.cluster_centers_

plt.figure(figsize = (8,8))

plt.scatter(df2['wifi_level'], df2['internal_temperature'],
            c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)

plt.xlabel("wifi_level")
plt.ylabel("internal_temperature")

plt.show()
```

```python
import pandas as pd

df = pd.read_csv("rawdata-mysql.csv")

df
```

```python
df2 = df.drop(columns=['ts','topic','chipid','board','firmware'])

df2
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2).fit(df2)
df2['kmean'] = kmeans.labels_

df2
```

```python
import pickle

pickle.dump(kmeans, open("esp32_classifier.pkl", "wb"))
```
