```python
import mysql.connector
```

```python
conn = mysql.connector.connect(user='user', password='password', auth_plugin='mysql_native_password',
                              host='vm_address', port='mysql_port',
                              database='database')
```

```python
print(conn)
```

```python
query = "SELECT * FROM mosquitto_register"

mycursor = conn.cursor()

mycursor.execute(query)

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
import pandas

df = pandas.read_sql_query(query, conn)

df
```

```python
mycursor.close()
conn.close()
```

```python
df = df.loc[df['topic'].isin(['29/status','31/status'])]

df = df.rename(columns={"timestamp": "ts"}, errors="raise")

df
```

```python
import json

df['payload'] = df['payload'].apply(lambda x: json.loads(x))

df
```

```python
from pandas import json_normalize

df = pandas.concat([df, json_normalize(df['payload'])], axis=1, join="inner")

df
```

```python
df = df.drop(columns=['payload'])

df
```

```python
df = df.drop(columns=['minfreeheap'])

df
```

```python
df['board'] = "esp32"

df
```

```python
def version(x):
    if (x == "31/status"):
        return("arcond-ir")
    else:
        return("dht-tempumid")

df['firmware'] = df['topic'].apply(lambda x: version(x))

df
```

```python
cols = ['chipid','freeheap','internal_temperature','wifi_level']
df[cols] = df[cols].apply(pandas.to_numeric, errors='coerce', axis=1)

df.dtypes
```
